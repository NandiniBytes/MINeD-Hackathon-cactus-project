```python
import fitz  # PyMuPDF
import json
import os
import re

# Function to extract text from a given PDF file
def extract_text_from_pdf(pdf_path):
    doc = fitz.open(pdf_path)
    text = ""

    for page in doc:
        text += page.get_text("text") + "\n\n"  # Preserve line breaks for structure

    return text.strip()

# Function to structure extracted text into JSON format
def convert_text_to_json(text, pdf_name):
    lines = text.split("\n")  # Split into lines
    json_data = {"title": pdf_name.replace(".pdf", ""), "sections": {}}

    current_section = None
    json_data["introduction"] = ""

    for line in lines:
        line = line.strip()
        
        # Identify section headings (Capitalized words or common section names)
        if re.match(r"^[A-Z ]{3,}$", line) or line in ["Abstract", "Introduction", "Conclusion", "Methodology", "Results", "Discussion"]:
            current_section = line.strip()
            json_data["sections"][current_section] = ""
        elif current_section:
            json_data["sections"][current_section] += line + " "
        else:
            json_data["introduction"] += line + " "

    return json_data

# Function to save JSON file
def save_json(json_data, output_folder):
    json_filename = json_data["title"].replace(" ", "_") + ".json"
    json_path = os.path.join(output_folder, json_filename)
    
    with open(json_path, "w", encoding="utf-8") as json_file:
        json.dump(json_data, json_file, indent=4)
    
    print(f"‚úÖ JSON saved at: {json_path}")

# Get folder path from user input
folder_path = input("Enter the full folder path containing PDFs: ").strip()

# Check if the folder exists
if os.path.exists(folder_path):
    pdf_files = [f for f in os.listdir(folder_path) if f.endswith(".pdf")]
    
    if not pdf_files:
        print("‚ùå No PDF files found in the folder.")
    else:
        for pdf_file in pdf_files:
            pdf_path = os.path.join(folder_path, pdf_file)
            extracted_text = extract_text_from_pdf(pdf_path)
            json_data = convert_text_to_json(extracted_text, pdf_file)
            save_json(json_data, folder_path)
        
        print("üéâ All PDFs have been converted to JSON successfully!")
else:
    print("‚ùå Error: Folder does not exist. Please enter a valid folder path.")
```

```python
import os
import json
from pptx import Presentation
from pptx.util import Inches

def create_presentation(json_file, output_folder, style="formal"):
    print(f"üîÑ Processing: {json_file}")
    
    try:
        # Read the JSON file
        with open(json_file, "r", encoding="utf-8") as file:
            data = json.load(file)

        # Create a PowerPoint presentation
        prs = Presentation()

        # Slide 1: Title Slide
        title_slide_layout = prs.slide_layouts[0]  # Title Slide
        slide = prs.slides.add_slide(title_slide_layout)
        title = slide.shapes.title
        subtitle = slide.placeholders[1]
        title.text = data.get("title", "Untitled Presentation")
        subtitle.text = "Generated Automatically"

        # Slide 2: Introduction
        if "introduction" in data:
            content_slide_layout = prs.slide_layouts[1]  # Title and Content Layout
            slide = prs.slides.add_slide(content_slide_layout)
            title = slide.shapes.title
            content = slide.placeholders[1]
            title.text = "Introduction"
            content.text = data["introduction"]

        # Slides for each section
        if "sections" in data:
            for section, text in data["sections"].items():
                slide = prs.slides.add_slide(content_slide_layout)
                title = slide.shapes.title
                content = slide.placeholders[1]
                title.text = section  # Section Name as Title
                content.text = text  # Section Content

        # Slide for Key Points
        if "key_points" in data:
            bullet_slide_layout = prs.slide_layouts[1]  # Title and Content Layout
            slide = prs.slides.add_slide(bullet_slide_layout)
            title = slide.shapes.title
            content = slide.placeholders[1]
            title.text = "Key Points"
            content.text = "\n".join(["‚Ä¢ " + point for point in data["key_points"]])

        # Save the presentation
        output_path = os.path.join(output_folder, os.path.basename(json_file).replace(".json", ".pptx"))
        prs.save(output_path)
        print(f"‚úÖ PPT successfully saved at {output_path}")

    except Exception as e:
        print(f"‚ùå Error processing {json_file}: {e}")

```

```python
import os
import json
from pptx import Presentation
from pptx.util import Inches

# Folder paths
json_folder = r"C:\Users\Admin\Desktop\uni\MINeD Hackathon\Sample PDFs"  # Change this to your actual path
ppt_folder = r"C:\Users\Admin\Desktop\uni\MINeD Hackathon\Generated PPTs"    # Change this to your actual path

# Ensure output folder exists
os.makedirs(ppt_folder, exist_ok=True)

# Function to create PowerPoint from JSON
def create_ppt_from_json(json_path, ppt_path):
    with open(json_path, "r", encoding="utf-8") as json_file:
        paper_data = json.load(json_file)

    prs = Presentation()

    # Title Slide
    slide_layout = prs.slide_layouts[0]  # Title Slide Layout
    slide = prs.slides.add_slide(slide_layout)
    title = slide.shapes.title
    subtitle = slide.placeholders[1]
    
    title.text = paper_data.get("title", "Research Paper Presentation")
    subtitle.text = "Auto-generated from JSON"

    # Add slides for each section
    for section, content in paper_data.get("sections", {}).items():
        slide_layout = prs.slide_layouts[1]  # Title & Content Layout
        slide = prs.slides.add_slide(slide_layout)
        title = slide.shapes.title
        content_box = slide.placeholders[1]

        title.text = section  # Section title
        content_box.text = content[:200] + "..." if len(content) > 200 else content  # First 200 chars

        # If section is too long, split it across multiple slides
        remaining_text = content[200:]
        while len(remaining_text) > 0:
            slide = prs.slides.add_slide(slide_layout)
            content_box = slide.placeholders[1]

            content_box.text = remaining_text[:300] + "..." if len(remaining_text) > 300 else remaining_text
            remaining_text = remaining_text[300:]

    # Save PPT
    prs.save(ppt_path)

# Process all JSON files in the folder
for json_file in os.listdir(json_folder):
    if json_file.endswith(".json"):
        json_path = os.path.join(json_folder, json_file)
        ppt_filename = json_file.replace(".json", ".pptx")
        ppt_path = os.path.join(ppt_folder, ppt_filename)

        create_ppt_from_json(json_path, ppt_path)

print("‚úÖ PPTs successfully created with structured slides!")
```
