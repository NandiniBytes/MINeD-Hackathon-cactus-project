```python
import requests
from google.colab import drive
import sys
import os
import pandas as pd
from rich.console import Console
from IPython.display import Image as IPImage, display
from diffusers import StableDiffusionPipeline
import torch

drive.mount('/content/drive')
console = Console()

# ✅ Setup Path
current_dir = os.path.dirname(os.path.abspath('/content/drive/MyDrive/MINed-Hackathon/graphical-abstract-generator/src/main.py'))
parent_dir = '/content/drive/MyDrive/MINed-Hackathon/graphical-abstract-generator'
sys.path.insert(0, parent_dir)

# ✅ Text Extraction
class Extractor:
    def extract_title(self, text):
        return text.split('\n')[0] if text else "Untitled Research Paper"

    def extract_key_findings(self, text):
        return ["Finding 1", "Finding 2"]  # Placeholder

    def extract_methods(self, text):
        return ["Method 1", "Method 2"]  # Placeholder

    def extract_conclusion(self, text):
        if not text:
            raise ValueError("Text cannot be empty")
        return "Conclusion"  # Placeholder

# ✅ Customization
class Customizer:
    def __init__(self):
        self.color_scheme = None
        self.layout = None
        self.design_type = None

    def set_color_scheme(self, scheme):
        self.color_scheme = scheme

    def set_layout(self, layout):
        self.layout = layout

    def set_design_type(self, design_type):
        self.design_type = design_type

# ✅ Visualization with Hugging Face API
class Visualizer:
    def __init__(self, api_token):
        self.api_token = api_token
        self.pipe = StableDiffusionPipeline.from_pretrained("runwayml/stable-diffusion-v1-5", torch_dtype=torch.float16)
        self.pipe.to("cuda" if torch.cuda.is_available() else "cpu")  # Use GPU if available


    def create_charts(self, key_findings):

      print("Creating charts for key findings...")

    def overlay_text(self, image_path, text):
      print("Overlaying text elements...")



    def generate_illustration(self, prompt):
        try:
            print("🚀 Generating illustration with Hugging Face...")

            # Generate image using Hugging Face API
            image = self.pipe(prompt).images[0]

            # Save the image locally
            image_path = "/content/drive/MyDrive/MINed-Hackathon/output/generated_image.png"
            image.save(image_path)

            print(f"✅ Image saved at {image_path}")
            display(IPImage(filename=image_path))
            return image_path

        except Exception as e:
            print(f"❌ Error generating illustration: {e}")
            return None

# ✅ Main Program
def main():
    print("🎨 Welcome to the Graphical Abstract Generator!")

    extractor = Extractor()
    try:
        research_paper_text = input("📜 Enter research paper text: ")
        title = extractor.extract_title(research_paper_text)
        key_findings = extractor.extract_key_findings(research_paper_text)
        methods = extractor.extract_methods(research_paper_text)
        conclusion = extractor.extract_conclusion(research_paper_text)
    except Exception as e:
        print(f"❌ Error during text extraction: {e}")
        return

    customizer = Customizer()
    try:
        color_scheme = input("🎨 Choose a color scheme (dark, light, custom): ").lower()
        layout = input("📐 Choose a layout style (vertical, horizontal, mind-map): ").lower()
        design_type = input("📄 Choose a design type (icon-based, text-heavy): ").lower()
        customizer.set_color_scheme(color_scheme)
        customizer.set_layout(layout)
        customizer.set_design_type(design_type)
    except Exception as e:
        print(f"❌ Error during customization: {e}")
        return

    # Hardcoded Hugging Face API token
    
    visualizer = Visualizer(api_token)

    try:
        # Example DataFrame for chart generation
        data = pd.DataFrame({
            'category': [f'Finding {i+1}' for i in range(len(key_findings))],
            'value': range(len(key_findings))
        })
        visualizer.create_charts(data)

        # Hugging Face prompt
        prompt = f"Create a scientific graphical abstract for research titled '{title}' with key findings: {', '.join(key_findings)}. Methods: {', '.join(methods)}. Conclusion: {conclusion}"

        # Generate illustration using Hugging Face
        image_path = visualizer.generate_illustration(prompt)

        if image_path:
            text_to_overlay = f"Title: {title}\nConclusion: {conclusion}"
            visualizer.overlay_text(image_path, text_to_overlay)
            print("✅ Graphical abstract generated successfully!")
            display(IPImage(filename=image_path))
        else:
            print("❌ Error: Could not generate illustration.")
    except Exception as e:
        print(f"❌ Error during visualization: {e}")
        return

if __name__ == "__main__":
    main()
```
